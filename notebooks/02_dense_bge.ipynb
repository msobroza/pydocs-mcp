# 02 — Dense embeddings (bge-small)

Single-vector **dense** retrieval: each chunk and the query are embedded with [`BAAI/bge-small-en-v1.5`](https://huggingface.co/BAAI/bge-small-en-v1.5) (384-dim, on-device via FastEmbed) and ranked by cosine similarity. No API key, no extra install.

Semantic: it can match a paraphrased description to the right function even with no shared keywords — watch the queries BM25 missed.

### Search (Python pipeline)

```python
from nb_helpers import make_searcher, load_queries, show_results

# Build the retrieval PIPELINE in Python for this method (no CLI search).
search = make_searcher("configs/dense_bge.yaml", cache_dir='.pydocs-cache/dense_bge')
queries = load_queries()
```

```python
for q in queries:
    hits = await search(q['query'], limit=5)
    show_results(q, hits)
```

### Takeaway
Dense recovers semantically-described needles BM25 can't. Quality is bounded by the small 384-dim model — `03_dense_qwen3` tries a stronger one.
