# 06 — Dense embeddings (gte-modernbert-base)

Single-vector **dense** retrieval with **Alibaba-NLP/gte-modernbert-base** (768-dim) — a ModernBERT-backbone general-purpose embedder with an **8192-token context**, served via **sentence-transformers** (torch). It's *symmetric* (no separate query instruction).

**Requires** `pip install -e '.[sentence-transformers]'` (already in the `.venv-li` kernel). Its embedder differs from bge-small, so this method uses its **own** index (own cache dir).

### Index

```python
import sys
# Index sample_repo for this method (CLI does the heavy ingestion wiring).
# --no-inspect = read source statically (don't import the package).
!{sys.executable} -m pydocs_mcp --config configs/dense_modernbert.yaml index sample_repo \
    --cache-dir .pydocs-cache/dense_modernbert --force --no-inspect 2>&1 \
    | grep -E "Project:|Done:|rror" || true
```

### Search (Python pipeline)

```python
from nb_helpers import make_searcher, load_queries, show_results

# Build the retrieval PIPELINE in Python for this method (no CLI search).
search = make_searcher("configs/dense_modernbert.yaml", cache_dir='.pydocs-cache/dense_modernbert')
queries = load_queries()
```

```python
for q in queries:
    hits = await search(q['query'], limit=5)
    show_results(q, hits)
```

### Takeaway
ModernBERT's long context + recent pretraining can sharpen recall on longer code chunks vs. bge-small, at a heavier model cost. Compare its hits to the other dense models.
