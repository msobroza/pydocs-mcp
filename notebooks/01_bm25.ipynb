# 01 — BM25 (keyword search)

**BM25** over SQLite **FTS5**: ranks chunks by weighted term overlap with the query. No embeddings, no API key — the zero-setup baseline.

Great at exact identifier/term matches; weak when the query *paraphrases* the code (no shared words). The default pipeline (`chunk_search`) is BM25, so we pass **no** `--config`.

### Search (Python pipeline)

```python
from nb_helpers import make_searcher, load_queries, show_results

# Build the retrieval PIPELINE in Python for this method (no CLI search).
search = make_searcher(None, cache_dir='.pydocs-cache/bm25')
queries = load_queries()
```

```python
for q in queries:
    hits = await search(q['query'], limit=5)
    show_results(q, hits)
```

### Takeaway
BM25 nails queries that share words with the target function and its docstring, but misses paraphrased descriptions with no lexical overlap. Compare the same queries under dense retrieval next.
