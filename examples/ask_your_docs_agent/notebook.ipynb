# Ask-your-docs agent (notebook)

Chat with a LangGraph ReAct agent grounded in your pydocs-mcp indexes.
Prereqs: `pip install -r requirements.txt`, `OPENAI_API_KEY` set (any
placeholder works for local servers), and a workspace of indexed bundles
(see README step 1). Set the four values below.


```python
WORKSPACE = "~/pydocs-index"  # directory of {name}_{hash}.db bundles
MODEL = "gpt-4o-mini"  # any OpenAI-API-compatible model
BASE_URL = None  # OpenAI-compatible endpoint, e.g. "http://localhost:8000/v1" for vLLM / Ollama / LiteLLM
PYDOCS_CONFIG = "configs/serve_cpu_openvino.yaml"  # or None for the default embedder
```

```python
from agent import ask, build_agent

agent, llm = await build_agent(WORKSPACE, model=MODEL, base_url=BASE_URL, pydocs_config=PYDOCS_CONFIG)
history = []
```

```python
print(await ask(agent, history, "How do I open a database pool?"))
```

```python
# Follow-up — 'it' resolves from the conversation history.
print(await ask(agent, history, "And what does it return?"))
```
